# Transectos de Propriedades

<a id='section_id2'></a>


<div class="alert alert-block alert-success">


<b> Algumas vezes podemos usar projeções que são como fatias perpendiculares à superficie da terra, conhecida como, seções transversais. As seções transversais nos permitem visualizar perfis da superficie da Terra de lado. Em geologia, as seções transversais são utilizadas para descrever estruturas geológicas no interior da terra. Em oceanografia, as seções transversais são uteis para ilustrar mudanças nas propriedades da água (temperatura, salinidade, transporte) no oceano.
<b>
    </div>


![texto do link](https://www.researchgate.net/profile/Lars_Rudstam/publication/274098836/figure/fig3/AS:294824386285580@1447303019569/Ocean-Data-View-generated-sections-to-40-m-for-corrected-chl-a-m-g-l-1-for-20-26.png)

![texto do link](https://www.researchgate.net/profile/Alexandre_Stegner/publication/260159835/figure/fig1/AS:340579423670279@1458211870671/Vertical-cross-section-of-the-potential-density-anomaly-along-the-transect-T1b-This.png)

Iremos utilizar a mesma base de dados de vários comandos utilizados na [**Aula 7**](https://github.com/calixtops/OEPY/blob/main/notebooks/OEPY%20-%20Aula%2009%20%20-%20Opendap%20-%20Basemap%20-%20Quiver%20-%20SODA%20.ipynb), o que vai adiantar bastante o nosso trabalho. Portanto, iremos fazer a importação do dado via OpenDAP e já realizar um <em>subset</em> para a região do Atlântico Sul.

```python
import xarray as xr
import cartopy as cp
import matplotlib.pyplot as plt
from cmocean import cm
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
url = 'http://apdrc.soest.hawaii.edu:80/dods/public_data/SODA/soda_3.3.1/monthly'
dado = xr.open_dataset(url)
dado = dado.sel(time=slice("2000-01-01","2015-01-01"),
          lat=slice(-50,5),lon=slice(300,360))
```

Temos o dado para a subamostrado para a região do AS, agora, temos que selecionar um ponto de latitude ou longitude que será a latitude de refrência para a nossao seção de propriedades. No caso, escolhi a latitude de -11. Note que ao selecionar a latitude que queremos, podemos utilizar o método "nearest" que vai encontrar o ponto de latitude mais próximo ao que passamos. 

```python
section = dado.sel(lon=slice(320,360))
section = dado.sel(lat = -11,method='nearest')
```

Para preparar o nosso plot, temos que fazer um Meshgrid com as coordenadas que vão orientar o eixo X e Y do meu gráfico. Como eu tenho 1 ponto de latitude, varios de longitude e algumas pronfundidades, os meus eixos X e Y gridados serão a Longitude e a Profundidade. 

```python
X,Z = np.meshgrid(dado.lon,-dado.lev)
```

## Velocidade meridional

<a id='section_id3'></a>


```python
section.v
```

Para a primeira seção que vamos contruir, escolheremos a variavel **v (meridional velocity [m/s])** 

Agora ficou facil de montar a nossa seção. Lembrando que temos alguns passos de tempo, portanto, vamos selecionar somente 1 para a nossa seção. Experimente fazer o plot para diferentes periodos de tempo, depois disso você pode produzir um gif ou uma animação de forma fácil. 

```python
plt.figure(figsize=(20,10))

ax = plt.gca()
ax.set_facecolor('0')


cf = plt.contourf(X,Z,section.v[2],
                  levels=np.arange(-.5,.5,0.001),
                  cmap = cm.balance)

c2 = plt.contour(X,Z,section.v[2],colors='k', 
                 linewidths=2
                )

clabels = plt.clabel(c2, fontsize=10,fmt='%1.2f',colors='k',use_clabeltext=True)
plt.colorbar(cf)

#plt.ylim([-5500,0])
```

## Densidade pontencial

<a id='section_id4'></a>


Podemos testar o nosso plot de seção com outros parametros como a Densidade Potencial.

```python
plt.figure(figsize=(20,10))

ax = plt.gca()
ax.set_facecolor('0')


cf = plt.contourf(X,Z,section.prho[1]-1000,
                  levels = np.arange(22,28,0.01),
                  cmap = cm.dense)

c2 = plt.contour(X,Z,section.prho[1]-1000,colors='w', 
                 linewidths=2,levels = np.arange(22,28,1)
                )

clabels = plt.clabel(c2, fontsize=10,fmt='%1.2f',colors='k',use_clabeltext=True)
plt.colorbar(cf)

#plt.ylim([-5500,0])
```

## Temperatura

<a id='section_id5'></a>


```python
plt.figure(figsize=(20,10))

ax = plt.gca()
ax.set_facecolor('0')


cf = plt.contourf(X,Z,section.temp[1],
                  levels = np.arange(0,32,0.1),
                  cmap = cm.thermal,extend = 'both')

c2 = plt.contour(X,Z,section.temp[1],colors='w', 
                 linewidths=2,levels = np.arange(0,32,4)
                )

clabels = plt.clabel(c2, fontsize=10,fmt='%1.0f',colors='k',use_clabeltext=True)
plt.colorbar(cf)

#plt.ylim([-5500,0])
```

## Salinidade 

<a id='section_id5'></a>


```python
plt.figure(figsize=(20,10))

ax = plt.gca()
ax.set_facecolor('0')


cf = plt.contourf(X,Z,section.salt[1],
                  levels = np.arange(34,37,0.01),
                  cmap = cm.haline,extend = 'both')

c2 = plt.contour(X,Z,section.salt[1],colors='w', 
                 linewidths=2,levels = np.arange(32,40,1)
                )

clabels = plt.clabel(c2, fontsize=10,fmt='%1.0f',colors='k',use_clabeltext=True)
plt.colorbar(cf)
```

# Diagrama de Hovmoller 

<a id='section_id8'></a>


<div class="alert alert-block alert-success">
<b>
O diagrama de Hovmöller foi inventado pelo meteorologista dinamarquês Ernest Aabo Hovmöller(1912-2008). Normalmente, o tempo é exibido nas ordenadas(eixo y) e geolocalização(lat ou lon) na abscissa(eixo x). Eles são comumente usados em meteorologia e oceanografia. 
    </b>
    </div>



Já que estamos manipulando o dado de todas as formas, utilizaremos novamente a seleção de dados para preparar um Diagrama de Hovmoller. Note que fizemos o mesmo fatiamento que foi realizado na seção.

```python
hov = dado.sel(lon=slice(320,360))
hov = dado.sel(lev=0,lat = -11,method='nearest')
```

Contruiremos o nosso diagrama com o tempo no eixo y e a longitude no eixo x.

```python
X,T = np.meshgrid(hov.lon,hov.time)
```

```python
plt.figure(figsize=(10,13))


CF = plt.contourf(X,T,hov.v,
                  cmap = cm.balance,
                  levels = np.arange(-.5,.6,0.01),
                  extend = 'both'
                 )


plt.contour(X,T,hov.v,colors = 'k')

plt.colorbar(CF)
plt.xlim([323.5,340])
```

# Séries Temporais

<a id='section_id8'></a>


<div class="alert alert-block alert-success">

Uma série temporal é uma coleção de observações feitas sequencialmente ao longo do tempo. Temos "observações" para 15 variaveis com 180 passos de tempo, portanto, podemos montar séries temporais com as variáveis que temos. Para isso, precisamos escolher um pontode lat e lon dentro do nosso dado e fazer o plot da propriedade variando no tempo
    </div>



```python
point = dado.sel(lat = -35,lon = 305,method='nearest')
```

```python
import seaborn as sns
sns.set()
```

```python
point.salt.isel(lev=0).plot()
```

```python
point.salt.isel(lev=0).plot()
point.temp.isel(lev=0).plot()
```

Podemos utilizar os conceitos apresentados na [**Aula 04**](https://github.com/calixtops/OEPY/blob/main/notebooks/OEPY%20-%20Aula%2004%20-%20Pandas%20%2B%20CTD.ipynb) para apresentar uma série com duas propriedades, agora, compartilhando o eixo y;

```python

fig, ax1 = plt.subplots(figsize=(20,7))
psalt = point.salt.isel(lev=0).plot(ax=ax1,color='y',label='Salt')
ax1.tick_params(axis='y', labelcolor='y')
ax1.set_ylabel('Salinity [PSU]',color = 'y')

ax2 = ax1.twinx()  

ax2.set_ylabel('Potencial Temperature [degc]',color='r') 
ptemp = point.temp.isel(lev=0).plot(ax=ax2,color='r',label='Temp')
ax2.tick_params(axis='y', labelcolor='r')

# Adcionando legenda para os dois eixos em um mesmo box
lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax2.legend(lines + lines2, labels + labels2, loc=0)
```
