<div class="alert alert-block alert-success">


<b> As estruturas de repetição são utilizadas para que comandos sejam executadas diversas vezes. Basicamente, seria a repetição de um bloco de código diversas vezes.

Há basicamente dois tipos de repetição:
<ul>
<li> <B> Repetição condicional: </B> executa um bloco de código enquanto uma condição lógica for verdadeira <mark><i>(while)</i></mark> </li>
<li> <B> Repetição contável: </B> executa um bloco de código um número predeterminado de vezes <mark><i>(for)</i></mark> </li>
<b>
    </div>

# 1. Instrução While


<a id='section_id2'></a>

<div class="alert alert-block alert-success">


<b> O comando <mark>(while)</mark> faz com que um conjunto de instruções seja executado enquanto uma condição é atendida. Quando o resultado dessa condição passa a ser falso, a execução do <mark>loop</mark> é interrompida.
<b>
    </div>

<div class="alert alert-block alert-success">


Em Python, não temos uma chave para fechar os blocos como temos em outras linguagens, como Matlab. Os blocos serão delimitados por uma identação visual, pela profundidade da identação em relação à margem esquerda do bloco de código. O mau uso da identação acarretará na não execução ou mal funcionamento do código.
    </div>

```python
contador = 0
while (contador < 5):
  print(contador)
  contador   = contador + 1
  print('Uma volta no bloco')
# Sem idententação não há repetição
```

Para esse exemplo, enquanto a variável contador, inicializada com 0, for menor do que 5, as instruções das linhas 3, 4 e 5 serão executadas.  A repetição condicional executa o bloco de instruções enquanto a condição for verdadeira.

Notem que foi utilizado uma "identação" para indicar o bloco de código a ser repetido. Em Python, para indicar o bloco de código pertencente ao while, devemos indentar o código, conforme demonstrado no exemplo.

A condição utilizada é uma expressão booleana que pode fazer uso de quaisquer operadores, podendo ser um ou mais comandos.

```python
contador<5
```

# 2. Instrução if 

<a id='section_id3'></a>


<div class="alert alert-block alert-success">


O **if** é uma estrutura de condição que permite avaliar uma expressão e, de acordo com seu resultado, executar um determinado bloco de código.
    </div>



No exemplo abaixo, usamos uma condição pré estabelecida para condicionar a execução de um dado bloco de código, que no caso, é uma mensagem.

```python
temperatura = 22

if temperatura < 24:
    print ('A temperatura amostrada é {}, o mínimo valor de temperatura para a região é 24. Rejeitando dado de temperatura'
           .format(temperatura))
```

##  if-else 

Vimos anteriormente como utilizar o if para executar uma ação caso uma condição seja atendida. No entanto, nenhum comportamento específico foi definido para o caso de a condição não ser satisfeita. Quando isso é necessário, precisamos utilizar o comando <B>else<B>.

```python
temperatura = 26 
if temperatura < 24:
    print ('A temperatura amostrada é {}, o mínimo valor de temperatura para a região é 24. Rejeitando dado de temperatura'
           .format(temperatura))
else:
    print ('O valor de temperatura está dentro dos limites aceitáveis.')
        
```

Dessa vez, caso a condição avaliada na linha 2 não seja atendida, definimos um fluxo alternativo que o código deve seguir. Ou seja, se a temperatura for menor do que o valor estipulado, o bloco abaixo da condição if será executada, caso não, o bloco abaixo do else será executado;

##  if-elif-else 

Adicionalmente, se existir mais de uma condição alternativa que precisa ser verificada, devemos utilizar o **elif** para avaliar as expressões intermediárias antes de usar o else, da seguinte forma:

```python
idade = 34
if idade < 12:
            print('criança')
elif idade < 18:
            print('adolescente')
elif idade < 60:
            print('adulto')
else:
            print('idoso')
```

<div class="alert alert-block alert-info">

Na linha 2 definimos a primeira condição (idade$<$12). Caso essa condição não seja atendida, o programa seguirá para a linha 4 e avaliará a próxima condição (elif), que se for verdadeira fará com que o bloco logo abaixo (a linha 5, nesse caso) seja executado. Caso essa condição ainda não seja atendida (elif), há uma outra alternativa na linha 6 que será avaliada e que fará com que o bloco logo abaixo seja executado se ela for atendida. Por fim, se nenhuma das condições for satisfeita, o programa seguirá para a linha 8, executando o que é definido pelo else.
    </div>

<b> Voltando ao While 

Uma estrutura de repetição é, muitas vezes, chamada de <B> loop </B> porque continua-se voltando ao início da instrução até que a condição se torna falsa; Se nao houver a condição falsa o loop será executado infinitamente, evite isso.

# 3. Repetição contável - Instrução For

<a id='section_id4'></a>


<div class="alert alert-block alert-success">


A instrução **for** é usada para iterar sobre os elementos de uma sequência (como uma string, tupla ou lista) ou outro objeto iterável:
    </div>

A sintaxe para execução da instrução For é bem semelhante às instruções while e if. No lugar de uma condição, temos uma lista,tupla ou string.

<div style='background-color: black'>

<p style='color: white'>
    for variavel in lista:<br>
    [identação]comandos
    </p>
    </div>

Enquanto percorremos a lista de valores, a variável indicada no for receberá, a cada iteração, um item da coleção. Assim, podemos executar algum processamento com esse elemento. No código abaixo percorremos a lista nomes e imprimimos cada elemento.

```python
nomes = ['Pedro', 'João', 'Leticia']

for n in nomes:
    print(n)
    print ('------------')
```

A variável definida na linha 1 é uma lista inicializada com uma sequência de valores do tipo string. A instrução **for** percorre todos esses elementos, um por vez e, em cada caso, atribui o valor do item à variável **n**, que é impressa em seguida. O resultado, então, é a impressão de todos os nomes contidos na lista.

```python
for letra in ['P','y','t','h','o','n']:
    print (letra)
```

É possível adicionar a instrução else ao final do for. Isso faz com que um bloco de código seja executado ao final da iteração, como mostra o exemplo a seguir:

```python
nomes = ['Pedro', 'João', 'Leticia']
for n in nomes:
     print(n)
else:
     print("Todos os nomes foram listados com sucesso")
```

**As instruções podem ser usadas em conjunto, respeitando a hierarquia da identação.**



```python
for time1 in ['Leao','Vozao','Ferrin','Icasa','Mengao']:
    for time2 in ['Leao','Vozao','Ferrin','Icasa','Mengao']:
        if time1 != time2:
            print(time1, 'x', time2)
```

# 4. Criando uma função - Def

<a id='section_id5'></a>


<div class="alert alert-block alert-success">

Uma função é um bloco de código que só é executado quando é chamado. Podemos passar parâmetros para as funções que utilizamos e também retornar dados de dentro da função. Em Python, uma função é definida usando o comando **def**.
    </div>

```python
def minha_funcao():
    print('Minha primeira função')
```

```python
minha_funcao()
```

Informações podem ser passadas para dentro da função na forma de argumentos. Os argumentos são especificados após nomearmos a função, entre os parênteses. Você pode adicionar quantos argumentos quiser, separando-os por vírgula.

```python
def mensagem(frase):
    print(frase)
```

```python
mensagem('Imprimindo uma frase qualquer')
```

Para retornar informações de dentro da função, usamos o comando **return**.

```python
def soma(a,b):
    c = a + b
    return c
```

```python
soma(4,6)
```

```python
c = soma(4,6)
```

```python
print(c)
```

# 5. Exemplos

<a id='section_id6'></a>


Como exemplo, faremos uma repetição para plotar as propriedades do CTD. A ideia do plot será a mesma utilizada na [**Aula 04**](https://github.com/calixtops/OEPY/blob/main/nb/OEPY%20-%20Aula%204%20-%20Pandas%20%2B%20CTD.ipynb).

Os arquivos a serem utilizados nesse exemplo estão na pasta **'dados/ctd/'**


Primeiro importamos as bibliotecas que iremos utilizar

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline  
```

<div class="alert alert-block alert-info">

**O módulo <mark> glob </mark> pode ser utilizado para encontrar caminhos de arquivos ou pastas de acordo com o padrão do Unix shell. Utilizaremos a função <mark> glob </mark> da biblioteca <mark> glob </mark> para criar uma lista com o nome de todos os arquivos da nossa pasta 'dados/ctd'.**    
    </div>



```python
from glob import glob
```

Usamos a sintaxe do shell para indicar que queremos todos(*) os arquivos .cnv

```python
lista_de_arquivos = glob('../dados/ctd/*.cnv')
```

Na váriavel **arquivo** temos uma lista com o caminho de todos os arquivos de ctd.

```python
lista_de_arquivos
```

Utilizando a instrução for, podemos percorrer toda essa lista.

```python
for arquivo in lista_de_arquivos:
    print('Arquivo a ser processado {0}'.format(arquivo))
    print('--------------Done--------------------------')
```

Copiando todos os comandos da [**Aula 04**](https://github.com/calixtops/OEPY/blob/main/nb/OEPY%20-%20Aula%204%20-%20Pandas%20%2B%20CTD.ipynb), podemos montar um plot e executar dentro do nosso loop.

Podemos utilizar algumas [**operações**](https://www.w3schools.com/python/python_ref_string.asp) com strings para dar nome e acertar o caminho que salvaremos a nossa figura. Confira o link!

Manipulando a string para montar um nome para a nossa figura. 

```python
lista_de_arquivos[0].replace('dados/ctd','imagens/aula_05').replace('.cnv','.png')
```

Manipulando a string para montar um titulo para nossa figura. 

```python
lista_de_arquivos[0].split('/')[-1].replace('.cnv','')
```

Executando loop com todos os arquivos da lista_de_arquivos com o codigo para realizar o plot.

```python
import seaborn as sns
sns.set()

for arquivo in lista_de_arquivos:
    
    fig_path = arquivo.replace('dados/ctd','imagens/aula_05').replace('.cnv','.png')
    str_title = arquivo.split('/')[-1].replace('.cnv','')


    dado = pd.read_csv(arquivo,
                           names=['prDM','t068C','c0S/m','depSM','sal00','sigma-t','density','svCM','dz/dtM','flag','nbin'],
                           skiprows=72,delim_whitespace= True)


    fig, ax1 = plt.subplots(figsize=(5,7))
    ax1.set_xlabel('Temperatura [ºC]',color = 'b')
    ax1.set_ylabel('Pressão [db]')
    ax1.plot(dado.t068C,-dado.prDM,'b') 
    ax1.tick_params(axis='x', labelcolor='b')


    ax2 = ax1.twiny()  # Inicializando um segundo eixo que vai compartilhar o mesmo a

    ax2.set_xlabel('Salinidade [PSU]',color='g') 
    ax2.plot(dado['sal00'],-dado.prDM,'g')
    ax2.tick_params(axis='x', labelcolor='g')
    plt.title('Plot de Temperatura e Salinidade com Eixo Compartilhado - Estação:{0}'.format(str_title))
    plt.savefig(fig_path,dpi=200)
```

Lembrando que não realizamos qualquer processamento nos dados de CTD, portanto, alguns ainda possuem os seus <em>bad values</em>. O objetivo dessa aula é exercitar as estruturas de repetição, portanto, não nos preocuparemos com a qualidade dos dados, por hora.

**Podemos utilizar a definição de funções para organizar melhor o nosso código. Vamos definir uma função para o plot de eixo compartilhado que fizemos acima.**

```python
def plot_2ax(fname):
    '''
    Função definida para plotar um gráfico de linha com eixo x compartilhado.
    
    input: fname='caminho para um arquivo .csv'
    
    '''
    
    fig_path = fname.replace('dados/ctd','imagens/aula_05').replace('.cnv','.png')
    str_title = fname.split('/')[-1].replace('.cnv','')


    dado = pd.read_csv(fname,
                           names=['prDM','t068C','c0S/m','depSM','sal00','sigma-t','density','svCM','dz/dtM','flag','nbin'],
                           skiprows=72,delim_whitespace= True)


    fig, ax1 = plt.subplots(figsize=(5,7))
    ax1.set_xlabel('Temperatura [ºC]',color = 'b')
    ax1.set_ylabel('Pressão [db]')
    ax1.plot(dado.t068C,-dado.prDM,'b') 
    ax1.tick_params(axis='x', labelcolor='b')


    ax2 = ax1.twiny()  # Inicializando um segundo eixo que vai compartilhar o mesmo a

    ax2.set_xlabel('Salinidade [PSU]',color='g') 
    ax2.plot(dado['sal00'],-dado.prDM,'g')
    ax2.tick_params(axis='x', labelcolor='g')
    plt.title('Plot de Temperatura e Salinidade com Eixo Compartilhado - Estação:{0}'.format(str_title))
    plt.savefig(fig_path,dpi=200)
```

Uma documentação pode ser passada junto com a sua função utilizando apostófros após a definição da função. É opcional mas é muito recomendado.

```python
help(plot_2ax)
```

Dessa forma, podemos realizar os nossos plots de forma mais rápida.

```python
for arquivos in lista_de_arquivos:
    plot_2ax(arquivos)
```
